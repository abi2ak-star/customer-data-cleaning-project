# Customer Data Cleaning with Python & Pandas

This project demonstrates how to clean and prepare messy customer data using Python and Pandas.

Tasks completed:
- Identified missing values
- Detected duplicate records
- Fixed invalid data types
- Handled missing values
- Generated summary statistics
- Exported a cleaned dataset

## 1. Import Libraries and Load Dataset

Import Pandas and load the messy customer dataset from Excel for analysis.

```python
import pandas as pd

df = pd.read_excel("messy_customer_data.xlsx")

print(df)
```

## 2. Explore Dataset Structure

Inspect data types, missing values and preview the dataset.

```python
print(df.info())
print(df.head())
```

## 3. Check Missing Values and Duplicate Records

Identify incomplete data and duplicate entries before cleaning.

```python
print(df.isnull().sum())
print(df.duplicated().sum())
```

## 4. Remove Duplicate Records

Remove identical duplicate rows to improve data quality.

```python
df = df.drop_duplicates()
```

## 5. Convert Invalid Data Types

Convert the Age column to numeric values and handle invalid entries.

```python
df["Age"] = pd.to_numeric(
    df["Age"],
    errors="coerce"
)
```

## 6. Handle Missing Values

Fill missing values using averages or placeholder values depending on the column type.

```python
df["Age"] = df["Age"].fillna(
    df["Age"].mean()
)

df["City"] = df["City"].fillna(
    "Unknown"
)

df["Purchase"] = df["Purchase"].fillna(
    df["Purchase"].mean()
)
```

## 7. Validate Cleaned Dataset

Check data types and confirm missing values were handled successfully.

```python
print(df.info())
print(df.isnull().sum())
```

## 8. Generate Summary Statistics

Review descriptive statistics to better understand the cleaned dataset.

```python
print(df.describe().round(2))
```

## 9. Export Clean Dataset

Save the cleaned dataset as a new Excel file.

```python
df.to_excel(
    "clean_customer_data.xlsx",
    index = False
)
```
